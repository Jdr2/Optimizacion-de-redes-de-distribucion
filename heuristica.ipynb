```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Coordenadas.csv', names=['Lat','Lon'])
df
```

```python
def calcular_distancia_haversine(longitud1, latitud1, longitud2, latitud2):

    longitud1 = np.radians(longitud1.astype(float))
    latitud1 = np.radians(latitud1.astype(float))
    longitud2 = np.radians(longitud2.astype(float))
    latitud2 = np.radians(latitud2.astype(float))

    radio_tierra = 6371

    diferencia_longitud = np.subtract(longitud2, longitud1)
    diferencia_latitud = np.subtract(latitud2, latitud1)

    a = np.add(np.power(np.sin(np.divide(diferencia_latitud, 2)), 2),
               np.multiply(np.cos(latitud1),
                           np.multiply(np.cos(latitud2),
                                       np.power(np.sin(np.divide(diferencia_longitud, 2)), 2))
                           )
              )
    c = np.multiply(2, np.arcsin(np.sqrt(a)))

    return c * radio_tierra
```

```python
def crear_matriz(distancias, num_nodos):
    matriz = [[0] * num_nodos for _ in range(num_nodos)]

    idx = 0
    for i in range(num_nodos):
        for j in range(num_nodos):
            matriz[i][j] = distancias[idx]
            idx += 1

    return matriz
```

```python
heuristica = []
for i in range(len(df['Lat'])):
    fila_heuristica = []
    for j in range(len(df['Lat'])):
        if i == j:
            dist = 0
        else:
            dist = calcular_distancia_haversine(df['Lon'][i], df['Lat'][i], df['Lon'][j], df['Lat'][j])
        fila_heuristica.append(dist)
    heuristica.append(fila_heuristica)
```

```python
direcciones = 25
headers = ['Depot']
for i in range(direcciones-1):
    headers.append(f'Cliente_{i+1}')
matrizH = pd.DataFrame(heuristica,columns=headers, index=headers)
matrizH
```
